```python
import pandas as pd

reviews = pd.read_csv('02_data.csv')
reviews
```

```python
# Uma função que temos usado intensamente até agora é a função value_counts ().
# Podemos replicar o que value_counts () faz fazendo o seguinte:
reviews.groupby('genero_musical').genero_musical.count()
```

```python
# faz a soma e mostra o dado predominando diente de outra coluna
reviews.groupby('genero_musical').estado.min()
```

```python
reviews.groupby('genero_musical').estado.agg([len, min, max])
```

```python
genero_estado = reviews.groupby(['estado', 'genero_musical']).linguagem.agg([len])
genero_estado.to_csv('genero_estado.csv')
```
