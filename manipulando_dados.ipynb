```python
# Primeira condição
pri_cond = df[df['idade'] > 50]
pri_cond
```

```python
# exibindo a quantidade de dados no df da primeira condição
len(pri_cond)
```

```python
# Segunda condição (maior que 50 e menor que 60)
seg_cond = df[(df['idade'] > 50) & (df['idade'] < 60)]
seg_cond
```

```python
# exibindo a quantidade de dados no df da segunda condição
len(seg_cond)
```

```python
# Ordenando dados
df_ordenado = df.sort_values(by=['idade'])
df_ordenado
```

```python
# Retirando coluna

df = df_ordenado.drop(columns=['email'])
df
```
