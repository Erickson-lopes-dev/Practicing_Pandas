```python
import pandas as pd
```

```python
# Syntax

df = pd.DataFrame(
    {
        'a': [4, 5, 6], 
        'b': [7, 8, 9],
        'c': [10, 12, 48]
    },
    index=[1,2,3],
    columns=['a', 'b', 'c']
)
df
```

```python
df = pd.DataFrame(
    [
        [1, 2, 3], 
        [4, 5, 6], 
        [7, 8, 9]
    ],
    index=[1, 2, 3], 
    columns=['A', 'B', 'C']
)
df
```

```python
df = pd.DataFrame(
    {
        'a': [4, 5, 6], 
        'b': [7, 8, 9],
        'c': [10, 12, 48]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('d', 1), ('d', 2), ('e', 2)
        ], 
        names=['n', 'v']
    )
)
df
```

```python
# Method Chaining
df = (pd.melt(df).rename(columns={'variable':'var', 'value': 'val'}).query('val >= 6'))
df
```

```python
# Reshaping Data 

df1 = pd.DataFrame({
        'a': [4, 5, 6], 
        'b': [7, 8, 9],
        'c': [10, 12, 48]
    },
    index=[1,2,3]
)
df1

df2 = pd.DataFrame({
        'a': [1, 2, 3], 
        'b': [11, 13, 14],
        'c': [20, 22, 44]
    },
    index=[4,5,6]
)
```

```python
# reunir colunas em linhas
pd.melt(df1)
```

```python
# concatena dfs
pd.concat([df1, df2])
```

```python
# espalhar linhas em colunas
df.pivot(columns='a', values='c')
```

```python
# anexar colunas de dataframes
pd.concat([df1, df2], axis=1)
```
