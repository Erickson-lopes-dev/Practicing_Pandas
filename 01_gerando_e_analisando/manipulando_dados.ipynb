```python
# Primeira condição
pri_cond = df[df['idade'] > 50]
pri_cond
```

```python
# exibindo a quantidade de dados no df da primeira condição
len(pri_cond)
```

```python
# Segunda condição (maior que 50 e menor que 60)
seg_cond = df[(df['idade'] > 50) & (df['idade'] < 60)]
seg_cond
```

```python
# exibindo a quantidade de dados no df da segunda condição
len(seg_cond)
```

```python
# Ordenando dados
df_ordenado = df.sort_values(by=['idade'])
df_ordenado
```

```python
# removendo coluna email
df2 = df_ordenado.drop(columns=['email'])
df2
```

```python
# Quantidade de programadores com genero_musical X

df2['genero_musical'].value_counts()
```

```python
# Exibindo quantidade 'linguagem por genero_musical

pd.pivot_table(df2[['linguagem', 'genero_musical']], index=['linguagem'], columns='genero_musical', aggfunc=[len]).T
```

```python
# captura quantidade de linguagens por estado
qtd_estado = pd.pivot_table(df2[['estado', 'linguagem']], index='linguagem', columns='estado', aggfunc=[len]).T

# substituindo dados nulos em 0
qtd_estado = qtd_estado.fillna(value=0)

qtd_estado
```
